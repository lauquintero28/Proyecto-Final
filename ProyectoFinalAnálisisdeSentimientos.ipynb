```python
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import sent_tokenize

# Descargar lexicon de VADER para análisis de sentimientos
nltk.download('vader_lexicon')  # Diccionario 70 idiomas.
sia = SentimentIntensityAnalyzer()

# Definir las categorías de polaridad
XC = np.array([-1, -0.8, -0.6, -0.4, -0.2, 0, 0.2, 0.4, 0.6, 0.8, 1])

# ------------------------------------------------------------------------------
# -------- Primer Informe - Discurso Presidencial Nicolas Maduro (2013) --------

blob = TextBlob(nxl)
tblob = blob.translate(from_lang='es', to='en')
nxl_tk = sent_tokenize(str(tblob))

nXC = np.zeros((11, 1))
Xtoks = []
Xpol = []
m1 = 0

for tokens in nxl_tk:
    print(tokens)
    scores = sia.polarity_scores(tokens)
    print(scores)
    Xtoks.append(tokens)
    Xpol.append(scores['compound'])
    polarity = scores['compound']

    d = np.abs(XC - polarity)
    pos = np.where(d == min(d))
    nXC[pos] = nXC[pos] + 1

# Se crea el informe en Excel para 2013
dfm = pd.DataFrame(np.column_stack((Xtoks, Xpol)))
dfm.columns = ['Tokens', 'Polaridad']
dfm.to_excel('Discurso_Nicolas_Maduro_2013.xlsx')

# ------------------------------------------------------------------------------
#---------Segundo Informe - Discurso Presidencial Nicolas Maduro (2014)---------

blob = TextBlob(nxl2)
tblob = blob.translate(from_lang='es', to='en')
nxl_tk2 = sent_tokenize(str(tblob))

nXC2 = np.zeros((11, 1))
Xtoks2 = []
Xpol2 = []
m2 = 0

for tokens2 in nxl_tk2:
    m2 = m2 + 1
    print(tokens2)
    scores = sia.polarity_scores(tokens2)
    print(scores)
    Xtoks2.append(tokens2)
    Xpol2.append(scores['compound'])

    polarity = scores['compound']

    d = np.abs(XC - polarity)
    pos = np.where(d == min(d))
    nXC2[pos] = nXC2[pos] + 1

# Se crea el informe en Excel para 2014
dfm2 = pd.DataFrame(np.column_stack((Xtoks2, Xpol2)))
dfm2.columns = ['Tokens', 'Polaridad']
dfm2.to_excel('Discurso_Nicolas_Maduro_2014.xlsx')

print("el número de comentarios es:", m2)

# ------------------------------------------------------------------------------
#---------Tercer Informe - Discurso Presidencial Nicolas Maduro (2019)----------

blob = TextBlob(nxl3)
tblob = blob.translate(from_lang='es', to='en')
nxl_tk3 = sent_tokenize(str(tblob))

nXC3 = np.zeros((11, 1))
Xtoks3 = []
Xpol3 = []
m3 = 0

for tokens3 in nxl_tk3:
    m3 = m3 + 1
    print(tokens3)
    scores = sia.polarity_scores(tokens3)
    print(scores)
    Xtoks3.append(tokens3)
    Xpol3.append(scores['compound'])

    polarity = scores['compound']

    d = np.abs(XC - polarity)
    pos = np.where(d == min(d))
    nXC3[pos] = nXC3[pos] + 1

# Se crea el informe en Excel para 2019
dfm3 = pd.DataFrame(np.column_stack((Xtoks3, Xpol3)))
dfm3.columns = ['Tokens', 'Polaridad']
dfm3.to_excel('Discurso_Nicolas_Maduro_2019.xlsx')

print("el número de comentarios es:", m3)

# ------------------------------------------------------------------------------
#---------Cuarto Informe - Discurso Presidencial Nicolas Maduro (2024)----------

blob = TextBlob(nxl4)
tblob = blob.translate(from_lang='es', to='en')
nxl_tk4 = sent_tokenize(str(tblob))

nXC4 = np.zeros((11, 1))
Xtoks4 = []
Xpol4 = []
m4 = 0

for tokens4 in nxl_tk4:
    m4 = m4 + 1
    print(tokens4)
    scores = sia.polarity_scores(tokens4)
    print(scores)
    Xtoks4.append(tokens4)
    Xpol4.append(scores['compound'])

    polarity = scores['compound']

    d = np.abs(XC - polarity)
    pos = np.where(d == min(d))
    nXC4[pos] = nXC4[pos] + 1

# Se crea el informe en Excel para 2024
dfm4 = pd.DataFrame(np.column_stack((Xtoks4, Xpol4)))
dfm4.columns = ['Tokens', 'Polaridad']
dfm4.to_excel('Discurso_Nicolas_Maduro_2024.xlsx')

print("el número de comentarios es:", m4)

# ------------------------------------------------------------------------------
#---------Graficación - Spider Plot Comparativo para los 4 años-----------------

labels = ['Amenaza', 'Pesimismo', 'Inestabilidad', 'Escepticismo', 'Indiferente', 'Neutro', 'Favorable', 'Optimismo', 'Convicción', 'Consolidación', 'Confianza']
angles = np.linspace(0, 2 * np.pi, 11, endpoint=False)
angles = np.concatenate((angles, [angles[0]]))

plt.figure(figsize=(8, 8))

colores = ['b', 'g', 'r', 'purple']

plt.polar(angles, np.concatenate((nXC, [nXC[0]])), marker='o', label='2013', color=colores[0], linewidth=2)
plt.polar(angles, np.concatenate((nXC2, [nXC2[0]])), marker='o', label='2014', color=colores[1], linewidth=2)
plt.polar(angles, np.concatenate((nXC3, [nXC3[0]])), marker='o', label='2019', color=colores[2], linewidth=2)
plt.polar(angles, np.concatenate((nXC4, [nXC4[0]])), marker='o', label='2024', color=colores[3], linewidth=2)

plt.fill(angles, np.concatenate((nXC, [nXC[0]])), alpha=0.25, color=colores[0])
plt.fill(angles, np.concatenate((nXC2, [nXC2[0]])), alpha=0.25, color=colores[1])
plt.fill(angles, np.concatenate((nXC3, [nXC3[0]])), alpha=0.25, color=colores[2])
plt.fill(angles, np.concatenate((nXC4, [nXC4[0]])), alpha=0.25, color=colores[3])

plt.thetagrids(angles[:-1] * 180 / np.pi, labels)
plt.title('Análisis de Sentimientos de los Discursos de Nicolás Maduro (2013-2024)', fontsize=16)
plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.05), fontsize=12)
plt.show()

print("Todos los informes Excel han sido generados correctamente.")
```
